```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('me.jpg')
```

```python
gray_filter = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
invert = cv2.bitwise_not(gray_filter)
```

```python
blur = cv2.GaussianBlur(invert, (31,31),0)
```

```python
inverted_blur = cv2.bitwise_not(blur)
```

```python
sketch_filter = cv2.divide(gray_filter, inverted_blur, scale = 256.0)
```

```python
output = cv2.cvtColor(sketch_filter, cv2.COLOR_BGR2RGB)
plt.imshow(output)
```

```python
sketch = cv2.imwrite('sketch.png', sketch_filter)
```
